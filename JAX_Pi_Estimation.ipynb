```python
import jax
import jax.numpy as jnp
from functools import partial

def in_circle(pt):
  return (pt[0]**2 + pt[1]**2) < 1

circ = jax.vmap(in_circle)

key = jax.random.PRNGKey(0)

@partial(jax.jit, static_argnums=(1,))
def pi(key, n, pts):
  inside = circ(pts)
  return 4 * jnp.mean(inside)

n_vals = [10 ** i for i in range(9)]

pis = jax.vmap(pi, in_axes = (None, None, 0))

for i in n_vals:
  key, subkey = jax.random.split(key)
  pts = jax.random.uniform(key, shape = (7, i, 2))
  estimates = pis(subkey, i, pts)
  print(f"{i} Points: {estimates} - Avg: {jnp.mean(estimates)}")
```
